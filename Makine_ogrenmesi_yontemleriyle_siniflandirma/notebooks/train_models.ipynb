```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# CSV'leri yükle
X_train = pd.read_csv("/content/drive/MyDrive/Derin_Ogrenme_Odev/Makine_ogrenmesi_yontemleriyle_siniflandirma/data/X_train.csv")
X_test = pd.read_csv("/content/drive/MyDrive/Derin_Ogrenme_Odev/Makine_ogrenmesi_yontemleriyle_siniflandirma/data/X_test.csv")
y_train = pd.read_csv("/content/drive/MyDrive/Derin_Ogrenme_Odev/Makine_ogrenmesi_yontemleriyle_siniflandirma/data/y_train.csv")
y_test = pd.read_csv("/content/drive/MyDrive/Derin_Ogrenme_Odev/Makine_ogrenmesi_yontemleriyle_siniflandirma/data/y_test.csv")

# y_train, y_test dizilerini düzleştir (pandas -> numpy)
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()

# Özellikleri ölçekle
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# ===== 1) Gerekli kütüphaneler =====
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# ===== 2) Logistic Regression Model Sınıfı =====
class LogisticRegressionModel:
    def __init__(self):
        self.model = LogisticRegression(max_iter=500)

    def train(self, X_train, y_train):
        self.model.fit(X_train, y_train)

    def evaluate(self, X_test, y_test):
        y_pred = self.model.predict(X_test)
        return {
            "Accuracy": accuracy_score(y_test, y_pred),
            "Precision": precision_score(y_test, y_pred),
            "Recall": recall_score(y_test, y_pred),
            "F1 Score": f1_score(y_test, y_pred)
        }

# ===== 3) Random Forest Model Sınıfı =====
class RandomForestModel:
    def __init__(self):
        self.model = RandomForestClassifier(n_estimators=200, max_depth=8, random_state=42)

    def train(self, X_train, y_train):
        self.model.fit(X_train, y_train)

    def evaluate(self, X_test, y_test):
        y_pred = self.model.predict(X_test)
        return {
            "Accuracy": accuracy_score(y_test, y_pred),
            "Precision": precision_score(y_test, y_pred),
            "Recall": recall_score(y_test, y_pred),
            "F1 Score": f1_score(y_test, y_pred)
        }

# ===== 4) SVM Model Sınıfı =====
class SVMModel:
    def __init__(self):
        self.model = SVC(kernel="rbf", C=1, gamma="scale")

    def train(self, X_train, y_train):
        self.model.fit(X_train, y_train)

    def evaluate(self, X_test, y_test):
        y_pred = self.model.predict(X_test)
        return {
            "Accuracy": accuracy_score(y_test, y_pred),
            "Precision": precision_score(y_test, y_pred),
            "Recall": recall_score(y_test, y_pred),
            "F1 Score": f1_score(y_test, y_pred)
        }

# ===== 5) Model örneklerini oluştur =====
log_model = LogisticRegressionModel()
rf_model = RandomForestModel()
svm_model = SVMModel()

# ===== 6) Modelleri eğit =====
log_model.train(X_train_scaled, y_train)
rf_model.train(X_train_scaled, y_train)
svm_model.train(X_train_scaled, y_train)

# ===== 7) Modelleri değerlendir ve metrikleri yazdır =====
print("===== Logistic Regression =====")
for key, value in log_model.evaluate(X_test_scaled, y_test).items():
    print(f"{key}: {value:.4f}")

print("\n===== Random Forest =====")
for key, value in rf_model.evaluate(X_test_scaled, y_test).items():
    print(f"{key}: {value:.4f}")

print("\n===== SVM =====")
for key, value in svm_model.evaluate(X_test_scaled, y_test).items():
    print(f"{key}: {value:.4f}")
```

```python
import pandas as pd

# Logistic Regression tahminleri
y_pred_log = log_model.model.predict(X_test_scaled)
pd.DataFrame({
    "y_true": y_test,
    "y_pred": y_pred_log
}).to_csv("results/logistic_regression_results.csv", index=False)

# Random Forest tahminleri
y_pred_rf = rf_model.model.predict(X_test_scaled)
pd.DataFrame({
    "y_true": y_test,
    "y_pred": y_pred_rf
}).to_csv("results/random_forest_results.csv", index=False)

# SVM tahminleri
y_pred_svm = svm_model.model.predict(X_test_scaled)
pd.DataFrame({
    "y_true": y_test,
    "y_pred": y_pred_svm
}).to_csv("results/svm_results.csv", index=False)
```

```python
import joblib

# Modelleri kaydet
joblib.dump(log_model.model, "results/logistic_regression_model.pkl")
joblib.dump(rf_model.model, "results/random_forest_model.pkl")
joblib.dump(svm_model.model, "results/svm_model.pkl")
```

```python
metrics = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest", "SVM"],
    "Accuracy": [
        log_model.evaluate(X_test_scaled, y_test)["Accuracy"],
        rf_model.evaluate(X_test_scaled, y_test)["Accuracy"],
        svm_model.evaluate(X_test_scaled, y_test)["Accuracy"]
    ],
    "Precision": [
        log_model.evaluate(X_test_scaled, y_test)["Precision"],
        rf_model.evaluate(X_test_scaled, y_test)["Precision"],
        svm_model.evaluate(X_test_scaled, y_test)["Precision"]
    ],
    "Recall": [
        log_model.evaluate(X_test_scaled, y_test)["Recall"],
        rf_model.evaluate(X_test_scaled, y_test)["Recall"],
        svm_model.evaluate(X_test_scaled, y_test)["Recall"]
    ],
    "F1 Score": [
        log_model.evaluate(X_test_scaled, y_test)["F1 Score"],
        rf_model.evaluate(X_test_scaled, y_test)["F1 Score"],
        svm_model.evaluate(X_test_scaled, y_test)["F1 Score"]
    ]
})

metrics.to_csv("results/models_metrics.csv", index=False)
```
